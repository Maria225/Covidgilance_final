```python
import pandas as pd
```

```python
health_xlsx = "County Health Rankings-Final.xlsx"
health_df = pd.read_excel(health_xlsx)
health_df.head()
```

```python
density_csv = "county_zipcode_true_ave.csv"
density_df = pd.read_csv(density_csv)
density_df = density_df.rename(columns={"county_fips": "fips"})
density_df.head()
```

```python
genderage_csv = "gender_age-Final.csv"
genderage_df = pd.read_csv(genderage_csv)
genderage_df.head()
```

```python
poverty_csv = "Poverty.csv"
poverty_df = pd.read_csv(poverty_csv)
poverty_df = poverty_df.rename(columns={"FIPStxt": "fips"})
poverty_df.head()
```

```python
merge12_df = pd.merge(health_df, density_df,  how='left', on = ['fips'])
merge12_df.head()
```

```python
merge123_df = pd.merge(merge12_df, genderage_df, how='left', on = ['fips'])
merge123_df.head()
```

```python
merge1234_df = pd.merge(merge123_df, poverty_df, how='left', on = ['fips'])
merge1234_df.head()
```

```python
merge1234_df = merge1234_df.drop(["state_x", "county_x", "county_name", "state_id", 
                                  "pop_area_est", "pop_area_est", "state_y", "county_y", "Stabr", "Area_name", 
                                  "Poverty, total"], axis=1)
merge1234_df
```

```python
covid_csv = "covid_cases_deaths.csv"
covid_df = pd.read_csv(covid_csv)
covid_df = covid_df.drop(["county", "state"], axis=1)
covid_df = covid_df.dropna(how='any')
covid_df
```

```python
totalmerged_df = pd.merge(covid_df, merge1234_df, how='left', on = ['fips'])
totalmerged_df = totalmerged_df.dropna(how='any')
totalmerged_df
```

```python
totalmerged_df.to_csv("totalmerged.csv", index=False, header=True)
```
