```python
import pandas as pd
```

```python
gender_age_xlsx = "Resources/gender_age.xlsx"
gender_age_df = pd.read_excel(gender_age_xlsx)
gender_age_df.head()
```

```python
fips_csv = "Resources/fips_county.csv"
fips_df = pd.read_csv(fips_csv)
fips_df.head()
```

```python
fips2_df = fips_df.drop(["state_abbr", "ssa"], axis=1)
fips2_df 
```

```python
gender_age2_df = pd.merge(gender_age_df, fips2_df,  how='left', on = ['state', 'county'])
gender_age2_df.head()
```

```python
gender_age2_df.to_csv ('Resources\clean_data\gender_age.csv', index = False, header=True)
```
